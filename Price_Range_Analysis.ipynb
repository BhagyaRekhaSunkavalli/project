```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_price_range_distribution(filepath):
    """
    Analyzes, visualizes, and calculates the percentage distribution of
    restaurant price ranges from a CSV file.

    Args:
        filepath (str): The path to the restaurant dataset CSV file.
    """
    try:
        # Load the dataset from the specified CSV file
        df = pd.read_csv(filepath)
        print("Dataset loaded successfully.\n")
    except FileNotFoundError:
        print(f"Error: The file '{filepath}' was not found.")
        return
    except Exception as e:
        print(f"An error occurred while reading the file: {e}")
        return

    print("--- Analyzing Price Range Distribution ---")

    # --- Task 1: Create a bar chart for price range distribution ---
    # Set a visually appealing style for the plot
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, ax = plt.subplots(figsize=(10, 6))

    # Count the occurrences of each price range and sort by the price range index (1, 2, 3, 4)
    price_range_counts = df['Price range'].value_counts().sort_index()

    # Create the bar chart using seaborn
    sns.barplot(
        x=price_range_counts.index,
        y=price_range_counts.values,
        palette="viridis",
        ax=ax,
        hue=price_range_counts.index, # Use hue to assign different colors from the palette
        legend=False # Hide the legend as the x-axis is clear
    )

    # Set titles and labels for clarity
    ax.set_title('Distribution of Price Ranges Among Restaurants', fontsize=16, fontweight='bold')
    ax.set_xlabel('Price Range Category', fontsize=12)
    ax.set_ylabel('Number of Restaurants', fontsize=12)

    # Add count labels on top of each bar to show the exact number
    # This makes the chart easier to read
    for container in ax.containers:
        ax.bar_label(container, fmt='%d', fontsize=10, color='black', padding=3)

    print("Displaying bar chart for price range distribution...")
    plt.tight_layout()
    plt.show() #

    # --- Task 2: Calculate the percentage of restaurants in each price range ---
    print("\n--- Percentage of Restaurants in Each Price Range ---")
    total_restaurants = len(df)
    # Calculate the percentage for each category
    price_range_percentage = (price_range_counts / total_restaurants) * 100

    # Print the results in a formatted way
    for price_range, percentage in price_range_percentage.items():
        count = price_range_counts[price_range]
        print(f"Price Range {price_range}: {count} restaurants ({percentage:.2f}%)")


if __name__ == "__main__":
    # The user uploaded 'Dataset .csv', so we use that path.
    analyze_price_range_distribution('/content/Dataset .csv')
```
